```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("accident.csv")
df.head()
```

```python
plt.scatter(df.Road_Type,df['Number_of_Vehicles'])
plt.xlabel("Road_Type")
plt.ylabel("Number_of_Vehicles")
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['Number_of_Vehicles']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head
```

```python
cluster_names = {
    0: 'Cluster 1',
    1: 'Cluster 2',
    2: 'Cluster 3',
    3: 'Cluster 4'
}
# Map cluster labels to cluster names
df['Cluster_Name'] = df['cluster'].map(cluster_names)
```

```python
km.cluster_centers_
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Number_of_Vehicles']])
    sse.append(km.inertia_)
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
